**DERET ARITMATIKA**

```python
# Rekursif Deret Aritmatika
def deret_aritmatika_recursive(a, d, n):
    if n == 1:
        return a
    return deret_aritmatika_recursive(a, d, n - 1) + d
```

```python
# Iteratif Deret Aritmatika
def deret_aritmatika_iterative(a, d, n):
    result = a
    for i in range(1, n):
        result += d
    return result
```

```python
import matplotlib.pyplot as plt

# Grafik untuk menyimpan data
n_values_aritmatika = []
recursive_times_aritmatika = []
iterative_times_aritmatika = []

# Fungsi untuk memperbarui grafik
def update_graph_aritmatika():
    plt.figure(figsize=(8, 6))
    plt.plot(n_values_aritmatika, recursive_times_aritmatika, label='Recursive', marker='o', linestyle='-')
    plt.plot(n_values_aritmatika, iterative_times_aritmatika, label='Iterative', marker='o', linestyle='-')
    plt.title('Performance Comparison: Recursive vs Iterative (Aritmatika)')
    plt.xlabel('Input (n)')
    plt.ylabel('Execution Time (seconds)')
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
from prettytable import PrettyTable

# Fungsi untuk mencetak tabel waktu eksekusi
def print_execution_table_aritmatika():
    table = PrettyTable()
    table.field_names = ["n", "Recursive Time (s)", "Iterative Time (s)"]
    min_len = min(len(n_values_aritmatika), len(recursive_times_aritmatika), len(iterative_times_aritmatika))
    for i in range(min_len):
        table.add_row([n_values_aritmatika[i], recursive_times_aritmatika[i], iterative_times_aritmatika[i]])
    print(table)
```

```python
import time
import matplotlib.pyplot as plt
from prettytable import PrettyTable

# Fungsi untuk menghitung deret aritmatika secara rekursif
def deret_aritmatika_recursive(a, d, n):
    if n == 1:
        return a
    return deret_aritmatika_recursive(a, d, n - 1) + d

# Fungsi untuk menghitung deret aritmatika secara iteratif
def deret_aritmatika_iterative(a, d, n):
    result = a
    for i in range(1, n):
        result += d
    return result

# Grafik untuk menyimpan data
n_values_aritmatika = []
recursive_times_aritmatika = []
iterative_times_aritmatika = []

# Fungsi untuk memperbarui grafik
def update_graph_aritmatika():
    plt.figure(figsize=(8, 6))
    plt.plot(n_values_aritmatika, recursive_times_aritmatika, label='Recursive', marker='o', linestyle='-')
    plt.plot(n_values_aritmatika, iterative_times_aritmatika, label='Iterative', marker='o', linestyle='-')
    plt.title('Performance Comparison: Recursive vs Iterative (Deret Aritmatika)')
    plt.xlabel('Input (n)')
    plt.ylabel('Execution Time (seconds)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Fungsi untuk mencetak tabel waktu eksekusi
def print_execution_table_aritmatika():
    table = PrettyTable()
    table.field_names = ["n", "Recursive Time (s)", "Iterative Time (s)"]

    min_len = min(len(n_values_aritmatika), len(recursive_times_aritmatika), len(iterative_times_aritmatika))
    for i in range(min_len):
        table.add_row([n_values_aritmatika[i], recursive_times_aritmatika[i], iterative_times_aritmatika[i]])

    print(table)

# Program utama untuk Deret Aritmatika
while True:
    try:
        # Meminta input berupa daftar angka yang dipisahkan koma
        n_values_input = input("Masukkan nilai n (atau ketik -1 untuk keluar): ")

        # Menghentikan program jika -1 dimasukkan
        if n_values_input == "-1":
            print("Program selesai. Terima kasih!")
            break

        # Mengubah input string menjadi daftar angka
        n_values_list = [int(x) for x in n_values_input.split(",")]

        # Angka awal dan beda deret aritmatika
        a = int(input("Masukkan nilai awal deret (a): "))
        d = int(input("Masukkan beda deret (d): "))

        for n in n_values_list:
            if n < 0:
                print(f"Nilai n = {n} tidak valid, harus positif!")
                continue

            n_values_aritmatika.append(n)

            # Ukur waktu eksekusi algoritma rekursif
            start_time = time.time()
            deret_aritmatika_recursive(a, d, n)  # rekursif
            recursive_times_aritmatika.append(time.time() - start_time)

            # Ukur waktu eksekusi algoritma iteratif
            start_time = time.time()
            deret_aritmatika_iterative(a, d, n)  # iteratif
            iterative_times_aritmatika.append(time.time() - start_time)

            # Cetak tabel waktu eksekusi
            print_execution_table_aritmatika()

            # Perbarui grafik
            update_graph_aritmatika()

    except ValueError:
        print("Masukkan nilai n yang valid!")
```

**DERET GEOMETRI**

```python
# Rekursif Deret Geometri
def deret_geometri_recursive(a, r, n):
    if n == 1:
        return a
    return deret_geometri_recursive(a, r, n - 1) * r
```

```python
# Iteratif Deret Geometri
def deret_geometri_iterative(a, r, n):
    result = a
    for i in range(1, n):
        result *= r
    return result
```

```python
# Fungsi untuk mencetak tabel waktu eksekusi Deret Geometri
def print_execution_table_geometri():
    table = PrettyTable()
    table.field_names = ["n", "Recursive Time (s)", "Iterative Time (s)"]
    min_len = min(len(n_values_geometri), len(recursive_times_geometri), len(iterative_times_geometri))
    for i in range(min_len):
        table.add_row([n_values_geometri[i], recursive_times_geometri[i], iterative_times_geometri[i]])
    print(table)
```

```python
import time
import matplotlib.pyplot as plt
from prettytable import PrettyTable

# Fungsi untuk menghitung deret geometri secara rekursif
def deret_geometri_recursive(a, r, n):
    if n == 1:
        return a
    return deret_geometri_recursive(a, r, n - 1) * r

# Fungsi untuk menghitung deret geometri secara iteratif
def deret_geometri_iterative(a, r, n):
    result = a
    for i in range(1, n):
        result *= r
    return result

# Grafik untuk menyimpan data
n_values_geometri = []
recursive_times_geometri = []
iterative_times_geometri = []

# Fungsi untuk memperbarui grafik
def update_graph_geometri():
    plt.figure(figsize=(8, 6))
    plt.plot(n_values_geometri, recursive_times_geometri, label='Recursive', marker='o', linestyle='-')
    plt.plot(n_values_geometri, iterative_times_geometri, label='Iterative', marker='o', linestyle='-')
    plt.title('Performance Comparison: Recursive vs Iterative (Deret Geometri)')
    plt.xlabel('Input (n)')
    plt.ylabel('Execution Time (seconds)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Fungsi untuk mencetak tabel waktu eksekusi
def print_execution_table_geometri():
    table = PrettyTable()
    table.field_names = ["n", "Recursive Time (s)", "Iterative Time (s)"]

    min_len = min(len(n_values_geometri), len(recursive_times_geometri), len(iterative_times_geometri))
    for i in range(min_len):
        table.add_row([n_values_geometri[i], recursive_times_geometri[i], iterative_times_geometri[i]])

    print(table)

# Program utama untuk Deret Geometri
while True:
    try:
        # Meminta input berupa daftar angka yang dipisahkan koma
        n_values_input = input("Masukkan nilai n (atau ketik -1 untuk keluar): ")

        # Menghentikan program jika -1 dimasukkan
        if n_values_input == "-1":
            print("Program selesai. Terima kasih!")
            break

        # Mengubah input string menjadi daftar angka
        n_values_list = [int(x) for x in n_values_input.split(",")]

        # Angka awal dan rasio untuk deret geometri
        a = int(input("Masukkan nilai awal deret (a): "))
        r = float(input("Masukkan rasio deret (r): "))

        for n in n_values_list:
            if n < 0:
                print(f"Nilai n = {n} tidak valid, harus positif!")
                continue

            n_values_geometri.append(n)

            # Ukur waktu eksekusi algoritma rekursif
            start_time = time.time()
            deret_geometri_recursive(a, r, n)  # rekursif
            recursive_times_geometri.append(time.time() - start_time)

            # Ukur waktu eksekusi algoritma iteratif
            start_time = time.time()
            deret_geometri_iterative(a, r, n)  # iteratif
            iterative_times_geometri.append(time.time() - start_time)

            # Cetak tabel waktu eksekusi
            print_execution_table_geometri()

            # Perbarui grafik
            update_graph_geometri()

    except ValueError:
        print("Masukkan nilai n yang valid!")
```
